```python
# Reload the full facility_data without nrows limit in order to obtain all NG rows

csv_file = 'EIA923_Schedules_2_3_4_5_M_12_2023_Final_Revision.csv'
facility_data = pd.read_csv(csv_file)

# Check how many rows are NG-based
ng_count = (facility_data['Reported\nFuel Type Code'] == 'NG').sum()
print(f"Rows with 'NG' fuel type: {ng_count}")

facility_data.head()
```

```python
# Clean the electricity_price dataframe for analysis
# 1. Filter for Year == 2023
# 2. Melt only the 'industrial cents per kilowatthour' columns to long format
# 3. Extract state name from column string
# 4. Create Grid_Price_per_MWh (dollars/MWh), value * 10
# 5. Save as clean_electric_prices

import re

df_2023 = electricity_price[electricity_price['Year'] == 2023]

# Select columns with 'industrial cents per kilowatthour'
cols_industrial = [col for col in df_2023.columns if 'industrial cents per kilowatthour' in col]

# Add 'Year' for melting
df_2023_industrial = df_2023[['Year'] + cols_industrial]

# Melt to long format
melted = df_2023_industrial.melt(id_vars=['Year'],
                                  value_vars=cols_industrial,
                                  var_name='Region_Sector',
                                  value_name='cents_per_kWh')
# Extract state name (before 'industrial')
def extract_state(colname):
    match = re.match(r'(.*) industrial cents per kilowatthour', colname)
    return match.group(1).strip() if match else colname

melted['State'] = melted['Region_Sector'].apply(extract_state)
# Convert from cents/kWh to $/MWh
melted['Grid_Price_per_MWh'] = melted['cents_per_kWh'] * 10

# Save cleaned dataframe
clean_electric_prices = melted[['Year', 'State', 'cents_per_kWh', 'Grid_Price_per_MWh']]
clean_electric_prices.head(20)
```

```python
# Filter clean_electric_prices to keep only actual U.S. states (no regions)

# List of 50 US states (full names)
us_state_names = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut',
    'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa',
    'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan',
    'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire',
    'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio',
    'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota',
    'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 
    'Wisconsin', 'Wyoming']

clean_electric_prices = clean_electric_prices[clean_electric_prices['State'].isin(us_state_names)].copy()

# Display for confirmation
clean_electric_prices.head(20)
```

```python
# Clean the fuel_price dataframe as requested
import re

# 1. Filter for Year == 2023
fuel_2023 = fuel_price[fuel_price['Year'] == 2023]

# 2. Melt to long format (exclude 'Year')
melted = fuel_2023.melt(id_vars=['Year'], var_name='Column_Desc', value_name='Gas_Cost_per_MMBtu')

# 3. Extract state name before 'dollars per mcf'
def extract_state(col):
    match = re.match(r'(.*) dollars per mcf', col)
    return match.group(1).strip() if match else col

melted['State'] = melted['Column_Desc'].apply(extract_state)

# 4. Save cleaned df
clean_fuel_cost = melted[['Year', 'State', 'Gas_Cost_per_MMBtu']]
clean_fuel_cost.head()
```

```python
# Filter clean_fuel_cost to keep only actual US states (no regions)
us_state_names = [
    'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut',
    'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa',
    'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan',
    'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire',
    'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio',
    'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota',
    'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia',
    'Wisconsin', 'Wyoming']

clean_fuel_cost = clean_fuel_cost[clean_fuel_cost['State'].isin(us_state_names)].copy()

# Display for confirmation
clean_fuel_cost.head(20)
```

```python
# Filter for Natural Gas plants and aggregate by plant
# This produces clean_plants as a summarized per-plant table
import pandas as pd

# Filter for NG plants
ng_plants = facility_data[facility_data['Reported\nFuel Type Code'] == 'NG'].copy()

# Ensure numeric columns (handle commas, missing, etc. if needed)
def parse_num(col):
    return pd.to_numeric(col.astype(str).str.replace(',', '').replace('.', '0'), errors='coerce')

ng_plants['Total Fuel Consumption\nMMBtu'] = pd.to_numeric(ng_plants['Total Fuel Consumption\nMMBtu'].astype(str).str.replace(',', ''), errors='coerce')
ng_plants['Net Generation\n(Megawatthours)'] = pd.to_numeric(ng_plants['Net Generation\n(Megawatthours)'].astype(str).str.replace(',', ''), errors='coerce')

# Group and sum
clean_plants = (
    ng_plants.groupby(['Plant Id', 'Plant Name', 'Plant State'], as_index=False)
             .agg({
                 'Total Fuel Consumption\nMMBtu': 'sum',
                 'Net Generation\n(Megawatthours)': 'sum'
             })
)

# Show first 10 rows
clean_plants.head(10)
```

```python
# Create a mapping from state abbreviations to full state names
state_map = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire',
    'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina',
    'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania',
    'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD': 'South Dakota', 'TN': 'Tennessee',
    'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington',
    'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming'
}

# Copy and standardize plant state to full name in clean_plants
def convert_state_abbr_to_name(row):
    abbr = row['Plant State']
    return state_map.get(abbr, abbr)

clean_plants_state = clean_plants.copy()
clean_plants_state['Plant State'] = clean_plants_state['Plant State'].apply(lambda x: state_map.get(x, x))

# Join with clean_fuel_cost on State
merged_1 = pd.merge(
    clean_plants_state,
    clean_fuel_cost.rename(columns={'State': 'Plant State'}),
    on='Plant State',
    how='inner'
)
# Join with clean_electric_prices on State
merged_master = pd.merge(
    merged_1,
    clean_electric_prices.rename(columns={'State': 'Plant State'}),
    on='Plant State',
    how='inner'
)

# Save as @master_analysis_table
master_analysis_table = merged_master

# Show first 10 rows
master_analysis_table.head(10)
```

```python
# Add an operational cost column: fuel consumption (MMBtu) * fuel cost ($/MMBtu)
master_analysis_table = master_analysis_table.copy()
master_analysis_table['Operational_Cost'] = master_analysis_table['Total Fuel Consumption\nMMBtu'] * master_analysis_table['Gas_Cost_per_MMBtu']
master_analysis_table.head(10)
```

```python
# Add Net Revenue as power generated times average grid price (per state)
master_analysis_table = master_analysis_table.copy()
master_analysis_table['Net Revenue'] = (
    master_analysis_table['Net Generation\n(Megawatthours)'] * master_analysis_table['Grid_Price_per_MWh']
)
master_analysis_table.head(10)
```

```python
# Create a clean final table from master_analysis_table as described
clean_final_plants = master_analysis_table[[
    'Plant Id',
    'Plant Name',
    'Plant State',
    'Net Generation\n(Megawatthours)',
    'Total Fuel Consumption\nMMBtu',
    'Net Revenue',
    'Operational_Cost',
    ]].copy()

# Rename 'Operational_Cost' column to 'Generation Cost'
clean_final_plants = clean_final_plants.rename(columns=
                                               {'Operational_Cost': 'Generation Cost', 
                                                'Net Generation\n(Megawatthours)' : 'Net Generation in MWh', 
                                                'Total Fuel Consumption\nMMBtu': 'Total Fuel Consumption in MMBtu'})

# Show the first 10 rows
clean_final_plants.head(10)
```

```python
# Convert generation cost to millions of dollars and round to two decimal places
clean_final_plants = clean_final_plants.copy()
clean_final_plants['Generation Cost (Million USD)'] = (clean_final_plants['Generation Cost'] / 1_000_000).round(2)
clean_final_plants = clean_final_plants.drop(columns=['Generation Cost'])

# Show the first 10 rows
clean_final_plants.head(10)
```

```python
# Convert net revenue to millions and round to two decimals, update column name
clean_final_plants = clean_final_plants.copy()
clean_final_plants['Net Revenue (Million USD)'] = (clean_final_plants['Net Revenue'] / 1_000_000).round(2)
clean_final_plants = clean_final_plants.drop(columns=['Net Revenue'])

# Show the first 10 rows
clean_final_plants.head(10)
```

```python
# Display columns to confirm correct naming
print(clean_final_plants.columns)

# Only select and show columns that definitely exist and fit the user's request
cols = ['Plant Id', 'Plant Name', 'Plant State', 'Net Generation in MWh', 'Total Fuel Consumption in MMBtu', 'Generation Cost (Million USD)', 'Net Revenue (Million USD)']

# Filter for columns that are actually present in the data
actually_present = [col for col in cols if col in clean_final_plants.columns]
clean_final_plants[actually_present].head(10)
```

```python
# Count plants with zero generation cost or zero revenues
n_zero_cost = (clean_final_plants['Generation Cost (Million USD)'] == 0).sum()
n_zero_revenue = (clean_final_plants['Net Revenue (Million USD)'] == 0).sum()

print(f"Plants with zero generation cost: {n_zero_cost}")
print(f"Plants with zero net revenue: {n_zero_revenue}")

# Show sample rows if they exist
if n_zero_cost > 0 or n_zero_revenue > 0:
    print("Sample rows with zero cost:")
    print(clean_final_plants[clean_final_plants['Generation Cost (Million USD)']==0].head())
    print("\nSample rows with zero revenue:")
    print(clean_final_plants[clean_final_plants['Net Revenue (Million USD)']==0].head())
```

```python
# Remove plants with zeroes in cost or revenue
good_plants = clean_final_plants[(clean_final_plants['Generation Cost (Million USD)'] > 0) &
                                  (clean_final_plants['Net Revenue (Million USD)'] > 0)].copy()
print(f"Rows after filtering: {len(good_plants)}")
good_plants.head(10)
```

```python
# Calculate Net Operating Profit and return the top 10 least profitable plants
plant_profit = good_plants.copy()
plant_profit['Net_Operating_Profit'] = plant_profit['Net Revenue (Million USD)'] - plant_profit['Generation Cost (Million USD)']

least_profitable = plant_profit.sort_values('Net_Operating_Profit').head(10)
least_profitable
```

```python
# Aggregate total net operating profit by Plant State
# Ensure Net_Operating_Profit exists in good_plants
good_plants = good_plants.copy()
good_plants['Net_Operating_Profit'] = (
    good_plants['Net Revenue (Million USD)'] - good_plants['Generation Cost (Million USD)']
)

state_profit = (
    good_plants.groupby('Plant State')['Net_Operating_Profit']
    .sum()
    .reset_index()
    .sort_values('Net_Operating_Profit', ascending=False)
)

print("Most profitable states:")
print(state_profit.head(10))
print("\nBiggest drains (least profitable states):")
print(state_profit.tail(10))
```

```python
# Compute Build_Cost_per_MWh and efficiency gap for all NG plants
plants_with_gap = master_analysis_table.copy()
plants_with_gap['Build_Cost_per_MWh'] = (
    plants_with_gap['Operational_Cost'] / plants_with_gap['Net Generation\n(Megawatthours)']
)
# Compute efficiency gap: build cost minus grid price
plants_with_gap['Efficiency_Gap'] = plants_with_gap['Build_Cost_per_MWh'] - plants_with_gap['Grid_Price_per_MWh']

# Filter for cases where build cost is higher than grid price
plants_gap_bad = plants_with_gap[plants_with_gap['Efficiency_Gap'] > 0]
# Top 10 biggest negative gaps (worst performing plants)
biggest_gap = plants_gap_bad.sort_values('Efficiency_Gap', ascending=False).head(10)

# Show relevant columns
biggest_gap[['Plant Id', 'Plant Name', 'Plant State', 'Build_Cost_per_MWh', 'Grid_Price_per_MWh', 'Efficiency_Gap']]
```

```python
# Convert operational cost and net revenue to millions in filtered_master
filtered_master = filtered_master.copy()
filtered_master['Generation Cost (Million USD)'] = (filtered_master['Operational_Cost'] / 1_000_000).round(2)
filtered_master['Net Revenue (Million USD)'] = (filtered_master['Net Revenue'] / 1_000_000).round(2)

# Show for confirmation
filtered_master[['Plant Id', 'Plant Name', 'Plant State', 'Generation Cost (Million USD)', 'Net Revenue (Million USD)']].head(10)
```

```python
# Calculate Net_Operating_Profit if not already present
filtered_master = filtered_master.copy()
filtered_master['Net_Operating_Profit'] = (
    filtered_master['Net Revenue (Million USD)'] - filtered_master['Generation Cost (Million USD)']
)

# Aggregate by Plant State
state_net_profit = (
    filtered_master.groupby('Plant State')['Net_Operating_Profit']
    .sum()
    .sort_values(ascending=False)
)

# Select top and bottom 5
top5 = state_net_profit.head(5)
bottom5 = state_net_profit.tail(5)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
# Plot top 5
sns.barplot(x=top5.values, y=top5.index, color='green', label='Top 5 most profitable')
# Plot bottom 5
y_offset = len(top5)
sns.barplot(x=bottom5.values, y=bottom5.index, color='red', label='Bottom 5 least profitable')
plt.xlabel('Net Operating Profit (USD)')
plt.title('Top 5 Most and Least Profitable States (Net Operating Profit)')
plt.legend(['Top 5 most profitable', 'Bottom 5 least profitable'])
plt.tight_layout()
plt.show()

# Also print state profit for clarity
print('Top 5 profitable states:')
print(top5)
print('\nLeast profitable states:')
print(bottom5)
```
